```python
#Thay đổi định dạng file test
import os
%cd /content/drive/MyDrive/CRAFT/tested_image
directory = os.listdir('/content/drive/MyDrive/CRAFT/tested_image')
count = 0
text_file = open("aaaa.txt","a")
for file in directory:
	if file.endswith(".txt"):
		count = count + 1
		with open(file) as f:
			content = f.readlines()
# Xóa bỏ các khoảng trăng hoặc các ký hiệu \n tại điểm cuối mỗi dòng
		content = [x.strip() for x in content] 
		for i in range(len(content)) :
			text_file.write("{0},{1}\n".format(file,content[i]))
   
print(count)
```
